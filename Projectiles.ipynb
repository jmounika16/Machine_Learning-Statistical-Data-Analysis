## Question 2: Carbon-14 Dating (5 points)

Carbon-14 dating is sometimes used to date fossils. It is known that the half-life of Carbon-14 is around 5730 years, meaning that every 5730 years, half of the remaining Carbon-14 decays. If a fossil still has $n$ percent of its carbon 14 remaining, then one can calculate its age (in years) according to the formula:

$$ Age= \frac{ln(\frac{n}{100})}{-0.693}*5730$$

&#9989;&nbsp; **a.** Construct a function which outputs a sample's age based on the percent, $n$, of Carbon-14 left in the sample. Call the function (using a value of your choice) and print the result using a clearly formatted and descriptive print statement.

&#9989;&nbsp; **b.** Construct a function which outputs the percent of Carbon-14 left in the sample based on its age. Call the function (using a value of your choice) and print the result using a clearly formatted and descriptive print statement. **HINT**: You should be able to rearrange the equation above to solve for $n$ as a function of $Age$ (you may need to remind yourself how to find the inverse of the natrual log, $ln$).

**NOTE**: Mathematical functions like, natural log ($ln$) or an exponential (i.e. $e^x$) can be called using the Python math library. You may need to look up what these functions are and don't forget to include "`import math`" in your code!

```python
import math

def print_age_from_percent(percent_of_carbon14_left):
    age = (math.log(percent_of_carbon14_left/100) * 5730) / (-0.693)
    print("Age of the fossil is " + str(round(age,3)) + " Years")

def print_percent_of_carbon_left_from_age(age):
    percent_left = math.exp((age* -0.693) / 5730 ) * 100
    print(str(percent_left) + " percent of carbon-14 left in the sample")

```

## Question 3: Fibonacci Sequence (5 points)

The Fibonacci sequence has the form $1,1,2,3,5,8, \dots$. Each successive number, after the first two, is found by adding the previous two numbers in the sequence. For example, the 7th number in the sequence is $5 + 8 = 13$.

**A possibly useful hint**: We might not have explicitly talked about it in class yet, but one really useful thing that Python allows you to do is index lists using *negative* numbers, which allows you to move backwards through the list as you increase the size of the negative number, starting with -1. That means that if I have the list `stuff = [1,2,3,4]`, I can do `stuff[-1]` and I'll get the last number, `4`. If I want the second to last number, I would do `stuff[-2]`.

&#9989;&nbsp; Your goal is to write a __*function*__ that computes $n$ terms in the sequence, where $n$ is larger than 1. You function should accept a __*value*__ $(n)$ (the number of terms in the sequence you wish to compute) and return the __*list*__ containing the first $n$ terms of the Fibonacci sequence. Test your function for $n = 4,5,6,7$ to ensure they match up with the sequence above.

You should get into the habit of never using code you find on the internet without citing it and understanding it.

```python
def generate_fibonacci_sequence(n):
    initial_sequence = [1,1]
    if n>1:
        for i in range(n):
            initial_sequence.append(initial_sequence[-1] + initial_sequence[-2])
    print(",".join(str(i) for i in initial_sequence[:n]))
generate_fibonacci_sequence(6)
            
```

## Question 4: Prime Numbers (5 points)

A prime number is a number larger than 1 that is only divisible by itself and 1. For example, 7 is a prime number as is 11.

&#9989;&nbsp;  **a.**  Write a function that outputs a list of primes that is less than a certain number $n$. Make sure your code only runs if $n$ is less than 1000 and that it prints a warning message to the user if the given value of $n$ is greater than 1000.

&#9989;&nbsp;  **b.** How many primes less than 100 are there? Include a print statement that answers this question using your function.

&#9989;&nbsp;  **c.** After running some tests with your code and thinking about the results, do you expect the fraction of primes in the interval $[0,n]$ to increase or decrease as $n$ gets larger? Why or why not?

```python
def check_prime(num):
    for i in range(2, num):
        if(num%i ==0):
            return False
    return True

def generate_prime(num):
    
    if num < 1000:
        lst = [1]
        for i in range(2,num):
            if check_prime(i):
                lst.append(i)
        print("There are " + str(len(lst)) + " primes under " + str(num))
        return lst

    else:
        print("Value is greater than 1000! Please enter a value less than 1000")
        
print(generate_prime(10))
print(generate_prime(50))
print(generate_prime(100))
print(generate_prime(200))
print(generate_prime(400))
print(generate_prime(800))
print(generate_prime(999))

```

```python
# https://docs.python.org/3/library/random.html#random.choice
from random import choice
import matplotlib.pyplot as plt

def update_random_walk_axis(ax):
    ax.xlabel("time")
    ax.ylabel("position")

def generate_random_walk(n):
    lst = [-1,1,0]
    walk = [0]
    for i in range(n):
        walk.append(walk[-1] + choice(lst))
    return walk
# print(generate_random_walk(100))

step_lst = [generate_random_walk(500) for i in range(3)]
#Referenced from:
# https://matplotlib.org/devdocs/gallery/subplots_axes_and_figures/subplots_demo.html
# https://stackoverflow.com/questions/6963035/pyplot-axes-labels-for-subplots
# https://www.geeksforgeeks.org/matplotlib-pyplot-tight_layout-in-python/
fig, ax_lst = plt.subplots(3)

for ax,step in zip(ax_lst,step_lst):
    ax.plot(step)
    ax.set_xlabel("time")
    ax.set_ylabel("position")
plt.tight_layout()
```

## Question 6: Projectiles (5 points)

An object is launched off a tower that is **100 meters tall**. We can model the height of the object at any point in time using this equation:

$$h(t) = h_{0}+ v_{0}t\sin(\theta) - \frac{1}{2}gt^2$$

where $h_{0}$ is the initial height, $v_0$ is the magnitude of the initial velocity *in meters per second*, $\theta$ is the angle that that the projectile is launched at *in radians*, $g$ is the acceleration due to gravity *in meters per seconds-squared*, and $t$ is the time *in seconds*.

&#9989;&nbsp; Using the equation above, run the following projectile motion simulations using the parameters below. Make *height* versus *time* plots for the first 15 seconds (in 1-second intervals) of the motion so that all three simulations are shown **on the same plot**.

**Hint**: you'll need to use the math module to access the $\sin$ and $\cos$ functions. You can also use the `math` module to convert from degrees to radians!

**a.** An object is thrown at an angle of 80 degrees upwards with a speed of 100 meters per second.

**b.** An object is thrown at an angle of 60 degrees upwards with a speed of 100 meters per second.

**c.** An object is thrown at an angle of 40 degrees upwards with a speed of 100 meters per second.

&#9989;&nbsp; After running your three simulations of projectile motion, review your results and use them to answer the following questions:

**d.** Approximately what angle should you launch your projectile at if your wanted it to reach a maximum height of roughly 400 meters? **Justify** your reasoning using your results.

**e.** For any of your three simulations, does the ball drop below the initial height of the tower within the 15 second window? If so, which one(s) and at roughly what time does this occur? (You should be able to use your plot to answer this question.)

```python
import math
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,20)
def get_projectile_height_at_t(v0, angle, t, h0=100):
    ht = h0 + (v0 * t * math.sin(angle * math.pi/180))  - (0.5* 9.81 * (t**2))
    return ht

def get_projectile_path(angle, upward_speed):
    return [get_projectile_height_at_t(upward_speed,angle,i) for i in range(16)]

a = get_projectile_path(80, 100)
b = get_projectile_path(60, 100)
c = get_projectile_path(40, 100)

simulations = [a,b,c]

fig, ax_lst = plt.subplots(3)

for ax,step,angle in zip(ax_lst,simulations,["80","60","40"]):
    ax.plot(step)
    ax.set_xlabel("time")
    ax.set_ylabel("height")
    ax.set_title("Angle=" + angle + ",max height = " + str(int(max(step))))
    ax.set_yticks([-100, 0, 100,200,300,400,500,600,700])
plt.tight_layout()
```
