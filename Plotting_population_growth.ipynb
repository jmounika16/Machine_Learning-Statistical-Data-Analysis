```python
# Put your code here
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,5)
def plot_data(dataset_path, delimiter=","):
    df = pd.read_csv("population_data.csv", sep=delimiter)
    df = df[['Time','Total Population']]
    df = df.sort_values('Time')
    df_pct_change = df.pct_change()
    years = df['Time']
    population = df_pct_change['Total Population']
    plt.plot(years,population)
    plt.xlabel("Years")
    plt.ylabel("Percentage change in Total population")
    plt.title("Growth of population over years")
plot_data('population_data.csv')
```
